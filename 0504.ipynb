```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df[:2]
```

```python
# 방법 1
df['target'] = df['target'].replace({0: 'setosa', 1: 'versicolor', 2: 'viginica'}) 
```

```python
# 방법 2
mapping = { 0 : 'setosa', 1: 'versicolor', 2: 'viginica'}
df['target'] = df['target'].map(mapping) 
```

```python
# 방법 3
df['target']= np.where(df['target']==0,'setosa', np.where(df['target']==1,'versicolor','viginica')) 
```

# 3교시 

```python
#setosa는 세모, versicolor는 네모, virginica는 동그라미로 표현
markers=['^', 's', 'o']

irisDF = df.copy()
irisDF.columns = [ 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
#setosa의 target 값은 0, versicolor는 1, virginica는 2. 각 target 별로 다른 shape으로 scatter plot 
for i, marker in enumerate(markers):
    x_axis_data = irisDF[irisDF['target']==i]['sepal_length']
    y_axis_data = irisDF[irisDF['target']==i]['sepal_width']
    plt.scatter(x_axis_data, y_axis_data, marker=marker,label=iris.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show() 
```
